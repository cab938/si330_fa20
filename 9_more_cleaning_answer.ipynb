```python
import pandas as pd
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States")[1]
df.head()
```

* Let's try and build a dataframe that has the president number, start and end date of term, name and birth/death years. I think there are likely questions like this on the citizenship test.

```python
df.columns=df.columns.droplevel(0)
```

```python
#val="April 30, 1789[d]–March 4, 1797"

def get_start(val):
    start=val.split("–")[0]
    return start

#get_start(val)
df["pres_start"]=df["Presidency[a].1"].apply(get_start)
```

```python
def get_start_by_row0(row):
    return row["Presidency[a].1"].split("–")[0]
df.apply(get_start_by_row0, axis=1)
```

```python
def get_start_by_row(row):
    row["presy_start"]=row["Presidency[a].1"].split("–")[0]
    return row
new_df=df.apply(get_start_by_row, axis=1)
new_df
```

```python
df['pres_name']=df["President.1"].str.replace(r'\d+.*','')
```

```python
df["President.1"].iloc[0]
```

```python
def get_years(row):
    import re
    #test='George Washington1732–1799(Lived: 67 years)[3][4][5]'
    test=row["President.1"]
    pattern=r'(?P<birth>\d\d\d\d).(?P<death>\d\d\d\d)'
    m=re.search(pattern,test)
    if m:
        m.groupdict()
        row["birth"]=m["birth"]
        row["death"]=m["death"]
        return row
    else:
        row["birth"]=None
        row["death"]=None
        return row
df.apply(get_years, axis=1)
```

```python
df[["real_birth","real_death"]]=df["President.1"].str.extract(r'(?P<birth>\d\d\d\d).(?P<death>\d\d\d\d)')
df
```
